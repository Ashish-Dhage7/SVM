```python
import pandas as pd 
import numpy as np 
import seaborn as sns
from sklearn.svm import SVC
from sklearn import preprocessing
```

```python
train_data = pd.read_csv("G:/ExcelR_Data/Data_Science_Course/Assignments/SVM/SalaryData_Train(1).csv")
train_data.head()
```

```python
test_data = pd.read_csv("G:/ExcelR_Data/Data_Science_Course/Assignments/SVM/SalaryData_Test(1).csv")
test_data.head()
```

```python
#Converting categorical data into numerical data
string_columns=["workclass","education","maritalstatus","occupation","relationship","race","sex","native","Salary"]
le = preprocessing.LabelEncoder()
for i in string_columns:
    train_data[i] = le.fit_transform(train_data[i])
    test_data[i] = le.fit_transform(test_data[i])
```

```python
train_data.head(2)
```

```python
train_data.shape
```

```python
test_data.head(2)
```

```python
#Splitting data into train and test
x_train = train_data.iloc[:,0:13]
y_train = train_data.iloc[:,13]
x_test  = test_data.iloc[:,0:13]
y_test  = test_data.iloc[:,13]
```

```python
#Build model where kernel is poly
model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
pred_test_poly = model_poly.predict(x_test)
```

```python
pd.crosstab(y_test,pred_test_poly)
```

```python
# classification_report
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_test_poly))
```

```python
#Build model where kernel is rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
pred_test_rbf = model_rbf.predict(x_test)
```

```python
pd.crosstab(y_test,pred_test_rbf)
```

```python
# classification_report
print(classification_report(y_test,pred_test_rbf))
```

```python
#Build model where kernel is sigmoid
model_sigmoid = SVC(kernel = "sigmoid")
model_sigmoid.fit(x_train,y_train)
pred_test_sigmoid = model_sigmoid.predict(x_test)
```

```python
pd.crosstab(y_test,pred_test_sigmoid)
```

```python
# classification_report
print(classification_report(y_test,pred_test_sigmoid))
```
